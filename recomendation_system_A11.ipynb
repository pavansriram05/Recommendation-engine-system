# Recommendation system

```python
anime_final = anime.set_index('anime_id')
anime_final
```

```python
def recommend(id):
    g= anime_final[anime_final.index==id].genre.item()

    M=anime_final[anime_final.index==id].name.item()

    movies = anime_final[anime_final['genre']==g].name.values

    movies=movies.tolist()

    movies.remove(M)

    if len(movies)!=0:
        print( list(movies))

    else:
        type = anime_final['type'].loc[id]

        movies_t= anime_final[(anime_final['type']==type) & (anime_final['rating']>=6)].name.tolist()
                                                                                                #list(anime_final[anime_final['type']==type].name)
        if anime_final['name'].loc[id] in movies_t:
            movies_t.remove(anime_final['name'].loc[3455])

        print(movies_t)
```
